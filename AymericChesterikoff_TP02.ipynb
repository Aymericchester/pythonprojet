**Exercise 8: Transformer and Pipeline with scikitlearn for Imputation processing**

In the previous exercise, we have decided to estimate the missing values using the mean estimator and fill missing
values with the same value. However the imputation is an estimation of the data with a defined strategy as
the median, the mean, variance, etc. Scikitlearn API provides an simple class to take care of missing values:
SimpleImputer. To use this class, we have to

• create a SimpleImputer instance by specifying the strategy

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy='mean')
```

• since the mean could be calculated only on numerical attributes, you need to create a copy of the data
without the text attributes. You call this copy as a origin Data Num

• Then you have to fit the mean on your origin Data Num

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy='mean')
Origin_Data_num=initialData.loc[:,'Price']
Origin_Data_num = Origin_Data_num.values.reshape(-1, 1)
my_imputer.fit(Origin_Data_num)
```

the fit() function stores the result into its statistic instance variable. At any moment you can reuse this
variable to repair new missing values. You can display this variable: imputer.statistics

• finaly You use the imputer to transform the numerical data:

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy='mean')
Origin_Data_num=initialData.loc[:,'Price']
Origin_Data_num = Origin_Data_num.values.reshape(-1, 1)
my_imputer.fit(Origin_Data_num)
print(my_imputer.statistics_)
my_imputer.transform(Origin_Data_num)
```

Be careful, the output of the transform function is a numpy array. You have to convert it to a dataframe:
