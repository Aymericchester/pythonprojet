```python
import numpy as np
import pandas as pd
flicker=pd.read_csv("/content/BL-Flickr-Images-Book-subset.csv",sep=",")
flicker
```

We display the columns and the null values in them

```python
flicker.isnull().sum()
```

Then, We delete the 4 columns that are empty 2880 nan out of 2880 rows

```python
columnstodelete = ["Corporate Author", "Corporate Contributors", "Former owner", "Engraver"]
flicker  = flicker.drop(columns=columnstodelete)
```

We then drop the rows with null values in them in order to keep only value without nan

```python
flicker = flicker.dropna()
flicker
```

Now, we need to make sure that the data is usable in the publication date column, that is to say that all the data is of the datetime type. We ensure that there will be no errors by using the error constraint.



```python
flicker['Date of Publication'] = pd.to_datetime(flicker["Date of Publication"], errors='coerce')
```

Then, we display the information about the cleaned dataset (only 143 value over 190 are datetime type in the publication data column).

```python
flicker.info()
```

Finally, we display a summary statistics of the date of publication column

```python
flicker['Date of Publication'].describe()
```
